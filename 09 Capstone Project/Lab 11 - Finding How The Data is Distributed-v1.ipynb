### Step 4: Analyze Key Columns


- Examine key columns such as `Employment`, `JobSat` (Job Satisfaction), and `YearsCodePro` (Professional Coding Experience).

```python
## Write your code here
df['JobSat'] = df['JobSat'].fillna(df['JobSat'].median())
df['JobSat'].isna().sum()

df['YearsCodePro'] = df['YearsCodePro'].fillna('0')
```

### Step 5: Visualize Job Satisfaction (Focus on JobSat)


- Create a pie chart or KDE plot to visualize the distribution of `JobSat`.

- Provide an interpretation of the plot, highlighting key trends in job satisfaction.


```python
## Write your code here
job_count = df['JobSat'].value_counts()

plt.figure(figsize = (14, 8))
plt.pie(job_count, labels=job_count.index, autopct='%1.1f%%')
plt.title('Job Satisfaction Distribution')
plt.tight_layout()
plt.show()
```

### Step 7: Analyze Remote Work Trends


- Visualize the distribution of RemoteWork by region using a grouped bar chart or heatmap.


```python
## Write your code here

df_remote = df[['Country', 'RemoteWork']].dropna()

remote_ct = pd.crosstab(df_remote['Country'], df_remote['RemoteWork'])

remote_ct.plot(kind='bar', figsize=(12, 8))
plt.xticks(rotation = 45)
plt.legend(title = 'Remote Work', loc = 'upper right')
plt.tight_layout()
plt.show()
```

### Step 9: Cross-tabulation Analysis (Employment vs. Education Level)


- Analyze the relationship between employment status (`Employment`) and education level (`EdLevel`).

- **Instruction**: Create a cross-tabulation using `pd.crosstab()` and visualize it with a stacked bar plot if possible.


```python
# Map long EdLevel labels to short names
edu_map = {
    'Primary/elementary school': 'Primary',
    'Secondary school (e.g. American high school, German Realschule or Gymnasium, etc.)': 'Secondary',
    'Some college/university study without earning a degree': 'Some College',
    'Associate degree (A.A., A.S., etc.)': 'Associate',
    'Bachelor’s degree (B.A., B.S., B.Eng., etc.)': 'Bachelor',
    'Master’s degree (M.A., M.S., M.Eng., MBA, etc.)': 'Master',
    'Professional degree (JD, MD, Ph.D, Ed.D, etc.)': 'Professional',
    'Something else': 'Other'
}
```

```python
## Write your code here
df_cross_analy = df[['Employment','EdLevel']].dropna()

# Map long EdLevel labels to short names
edu_map = {
    'Primary/elementary school': 'Primary',
    'Secondary school (e.g. American high school, German Realschule or Gymnasium, etc.)': 'Secondary',
    'Some college/university study without earning a degree': 'Some College',
    'Associate degree (A.A., A.S., etc.)': 'Associate',
    'Bachelor’s degree (B.A., B.S., B.Eng., etc.)': 'Bachelor',
    'Master’s degree (M.A., M.S., M.Eng., MBA, etc.)': 'Master',
    'Professional degree (JD, MD, Ph.D, Ed.D, etc.)': 'Professional',
    'Something else': 'Other'
}

df_cross_analy['EdLevel_short'] = df_cross_analy['EdLevel'].map(edu_map)

employment_map = {
    "Independent contractor, freelancer, or self-employed": "Freelance",
    "Full-time employee, including contractor": "Full-time",
    "Part-time employee": "Part-time",
    "Not employed, but looking for work": "Unemployed",
    "Not employed, and not looking for work": "Not Looking",
    "Retired": "Retired",
    "Student": "Student",
    "Currently not employed": "Other"
}

df_cross_analy['Employment_short'] = df_cross_analy['Employment'].map(employment_map)


df_crosstab_short = pd.crosstab(df_cross_analy['Employment_short'], df_cross_analy['EdLevel_short'], normalize='index') * 100
```

```python
df_crosstab_short.plot(kind='bar', figsize=(10, 6), colormap='tab20')
plt.xlabel("Percentage of Respondents")
plt.ylabel("Employment Type")
plt.legend(title="Education Level", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation = 0)
plt.tight_layout()
plt.show()
```
